```python
def fixTime(energy):
    #Splitting Date and Time apart
    dateTime = energy['Forecast Transaction Date'].str.split(' ', expand=True)
    #energy = energy.drop(['MCR'],1)
    #energy = energy.drop(['MC'],1)
    energy = energy.drop('Forecast Transaction Date',1)
    
    mapping={dateTime.columns[0]: 'Date', dateTime.columns[1]: 'Time Start'}
    dateTime = dateTime.rename(columns=mapping)
           
    #Seperating Hours and Minutes
    hourNum = dateTime['Time Start'].str.split(':', expand=True)
    mapping={hourNum.columns[0]: 'Hour (Ending)'}
    hourNum = hourNum.rename(columns=mapping)
    hourNum['Hour (Ending)'] = hourNum['Hour (Ending)'].astype(int)
    hourNum = hourNum['Hour (Ending)'] + 1
    dateTime = dateTime.drop(['Time Start'],1)
        
    #Putting everything back together
    energy = pd.concat([dateTime, hourNum, energy], axis=1)
    
    
   # energy = pd.concat([hourNum,energy], axis=1)
    #energy = energy.drop('Time',1)
   # energy = energy[['Date', 'Hour (Ending)', 'Min', 'Most Likely', 'Max']]
    
    #Differentiating the Mins/Maxs between power sources
    #mapping={energy.columns[2]: power +' Min', energy.columns[3]: power + ' Most Likely', energy.columns[4]: power + ' Max' }
    #energy = energy.rename(columns=mapping)
    
    return energy

#Use concatenation to combine two columns into one Use the syntax 
#DataFrame["new_column"] = DataFrame["column1"] + DataFrame["column2"] to combine two DataFrame columns into one.
```

```python
# Preping the data to Merge together/Changing time to Hour ending
def dataPrep(energy, maxCap):
    #Splitting Date and Time apart
    dateTime = energy['Forecast Transaction Date'].str.split(' ', expand=True)
    energy = energy.drop(maxCap,1)
    energy = energy.drop('Forecast Transaction Date',1)

    energy = pd.concat([dateTime,energy], axis=1)

    mapping={energy.columns[0]: 'Date', energy.columns[1]: 'Time' }
    energy = energy.rename(columns=mapping)
    
    #Seperating Hours and Minutes
    hourNum = energy['Time'].str.split(':', expand=True)
    mapping={hourNum.columns[0]: 'Hour (Ending)'}
    hourNum = hourNum.rename(columns=mapping)
    hourNum['Hour (Ending)'] = hourNum['Hour (Ending)'].astype(int)
    hourNum = hourNum['Hour (Ending)'] + 1

    #Putting everything back together in the Correct Order
    energy = pd.concat([hourNum,energy], axis=1)
    energy = energy.drop('Time',1)
    energy = energy[['Date', 'Hour (Ending)', 'Min', 'Most Likely', 'Max']]
    
    #Differentiating the Mins/Maxs between power sources
    #mapping={energy.columns[2]: power +' Min', energy.columns[3]: power + ' Most Likely', energy.columns[4]: power + ' Max' }
    #energy = energy.rename(columns=mapping)
    
    return energy
```
